```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv('kyphosis.csv')
```

```python
df.head()
```

```python
x = df.drop(['Kyphosis'], axis = 1)
y = df['Kyphosis']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_model = DecisionTreeClassifier()
dt_model.fit(x_train, y_train)
dt_pred = dt_model.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, dt_pred))
```

```python
rf_model = RandomForestClassifier()
rf_model.fit(x_train, y_train)
rf_pred = rf_model.predict(x_test)
```

```python
print(classification_report(y_test, rf_pred))
```

```python
df = pd.read_csv('loan_data.csv')
```

```python
df.head()
```

```python
categorical_column = ['purpose']
```

```python
df = pd.get_dummies(columns = categorical_column, data = df, drop_first = True)
```

```python
df.head()
```

```python
x = df.drop(['not.fully.paid'], axis = 1)
y = df['not.fully.paid']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.25, random_state = 42)
```

```python
dt_model = DecisionTreeClassifier()
dt_model.fit(x_train, y_train)
dt_pred = dt_model.predict(x_test)
```

```python
print(classification_report(y_test, dt_pred))
```

```python
rf_model = RandomForestClassifier()
rf_model.fit(x_train, y_train)
rf_pred = rf_model.predict(x_test)
```

```python
print(classification_report(y_test, rf_pred))
```

```python
df = pd.read_csv('XGBOOST_Data.csv')
```

```python
df.head()
```

```python
df.drop(columns = ['Unnamed: 0', 'id'], axis = 1, inplace = True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
```

```python
df['Gender'] = encoder.fit_transform(df['Gender'])
```

```python
df['Customer Type'] = encoder.fit_transform(df['Customer Type'])
```

```python
df['Type of Travel'] = encoder.fit_transform(df['Type of Travel'])
```

```python
df['Class'] = encoder.fit_transform(df['Class'])
```

```python
df.columns
```

```python
df.head()
```

```python
x = df.drop(['satisfaction'], axis = 1)
```

```python
y = df['satisfaction']
```

```python
y = encoder.fit_transform(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
```

```python
from xgboost import XGBClassifier
```

```python
model = XGBClassifier()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
model = AdaBoostClassifier()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
print(classification_report(y_test, y_pred))
```
