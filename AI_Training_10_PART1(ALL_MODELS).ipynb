```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
```

```python
data = pd.read_csv('data.csv')
```

```python
data.head()
```

```python
data = data.drop(['Unnamed: 32', 'id'], axis=1)
```

```python
le = LabelEncoder()
data['diagnosis'] = le.fit_transform(data['diagnosis'])
```

```python
X = data.drop('diagnosis', axis=1)
y = data['diagnosis']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model_results = {}
```

```python
log_reg = LogisticRegression()
log_reg.fit(X_train_scaled, y_train)
y_pred_log_reg = log_reg.predict(X_test_scaled)
model_results['Logistic Regression'] = accuracy_score(y_test, y_pred_log_reg)
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train_scaled, y_train)
y_pred_knn = knn.predict(X_test_scaled)
model_results['KNN'] = accuracy_score(y_test, y_pred_knn)
```

```python
svm = SVC()
svm.fit(X_train_scaled, y_train)
y_pred_svm = svm.predict(X_test_scaled)
model_results['SVM'] = accuracy_score(y_test, y_pred_svm)
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)
model_results['Decision Tree'] = accuracy_score(y_test, y_pred_dt)
```

```python
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
model_results['Random Forest'] = accuracy_score(y_test, y_pred_rf)
```

```python
ada_boost = AdaBoostClassifier(n_estimators=100, random_state=42)
ada_boost.fit(X_train, y_train)
y_pred_ada = ada_boost.predict(X_test)
model_results['AdaBoost'] = accuracy_score(y_test, y_pred_ada)
```

```python
for model, accuracy in model_results.items():
    print(f"{model}: Accuracy = {accuracy:.4f}")
```

```python
import time

# Dictionary to store time taken for each model
model_times = {}

log_reg = LogisticRegression(random_state=42)
start_time = time.time()
log_reg.fit(X_train_scaled, y_train)
y_pred_log_reg = log_reg.predict(X_test_scaled)
model_times['Logistic Regression'] = time.time() - start_time

knn = KNeighborsClassifier()
start_time = time.time()
knn.fit(X_train_scaled, y_train)
y_pred_knn = knn.predict(X_test_scaled)
model_times['KNN'] = time.time() - start_time

svm = SVC(random_state=42)
start_time = time.time()
svm.fit(X_train_scaled, y_train)
y_pred_svm = svm.predict(X_test_scaled)
model_times['SVM'] = time.time() - start_time

dt = DecisionTreeClassifier(random_state=42)
start_time = time.time()
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)
model_times['Decision Tree'] = time.time() - start_time

rf = RandomForestClassifier(random_state=42)
start_time = time.time()
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
model_times['Random Forest'] = time.time() - start_time

ada_boost = AdaBoostClassifier(n_estimators=100, random_state=42)
start_time = time.time()
ada_boost.fit(X_train, y_train)
y_pred_ada = ada_boost.predict(X_test)
model_times['AdaBoost'] = time.time() - start_time

# Display the time taken for each model
print(model_times)
```
