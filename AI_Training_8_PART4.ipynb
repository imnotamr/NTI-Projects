```python
data = pd.read_csv('Copy of titanic.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data = data.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)
data.head()
```

```python
data.isnull().sum()
```

```python
data['Age'].fillna(data['Age'].mean(), inplace=True)
data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder_sex = LabelEncoder()
data['Sex'] = label_encoder_sex.fit_transform(data['Sex'])
```

```python
label_encoder_embarked = LabelEncoder()
data['Embarked'] = label_encoder_embarked.fit_transform(data['Embarked'])
```

```python
data.head()
```

```python
x = data.drop('Survived', axis=1)
y = data['Survived']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
logistic_model = LogisticRegression()
logistic_model.fit(x_train, y_train)
```

```python
y_pred = logistic_model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
logistic_accuracy = accuracy_score(y_test, y_pred)
logistic_accuracy
```

```python
knn_model = KNeighborsClassifier()
```

```python
knn_model = KNeighborsClassifier()
params = {
    'n_neighbors' : list(range(1,25)),
    'p' : [1,2],
    'weights' : ['uniform', 'distance'],
    'metric' : ['minkowski', 'manhattan'],
    'algorithm' : ['auto', 'ball_tree', 'kd_tree', 'brute']
}
```

```python
from sklearn.model_selection import GridSearchCV
gs_model = GridSearchCV(knn_model, params)
gs_model.fit(x_train, y_train)
```

```python
gs_model.best_params_
```

```python
model = KNeighborsClassifier(n_neighbors = 6 , p = 1 , weights = 'distance' , metric = 'minkowski' , algorithm = 'auto')
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
knn_accuracy = accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
print(f"Logistic Regression Accuracy: {logistic_accuracy}")
print(f"KNN Accuracy: {knn_accuracy}")
```
